```python
import pandas as pd
import numpy as np
```

```python
djai=pd.read_csv('data/DJAI.csv')
drs=pd.read_csv('data/DRSRET100S.csv')
emp=pd.read_csv('data/Emp_pop_per.csv')
fdi=pd.read_csv('data/fdi.csv')
fed_spending=pd.read_csv('data/Federal Infrastructure Spending.csv')
gdp=pd.read_csv('data/GDP.csv')
hai=spending=pd.read_csv('data/HAI.csv')
price=pd.read_csv('data/Home Price Index.csv')
inf=pd.read_csv('data/inflation_rate.csv')
mort=pd.read_csv('data/MORTGAGE30US.csv')
inc=pd.read_csv('data/Per capita.csv')
permit=pd.read_csv('data/PERMIT.csv')
crime=pd.read_csv('data/Rate-of-Violent-Crime-Offenses-by-Population.csv')
```

```python
price=price.merge(emp,how='left')

price['YEAR']=pd.DatetimeIndex(price['DATE']).year
price['MONTH']=pd.DatetimeIndex(price['DATE']).month
```

```python
gdp=gdp.merge(fdi,how='left')
gdp=gdp.merge(drs,how='left')
gdp=gdp.merge(inc,how='left')
```

```python
gdp['YEAR']=pd.DatetimeIndex(gdp['DATE']).year
gdp['MONTH']=pd.DatetimeIndex(gdp['DATE']).month

price=price.merge(gdp,how='left')
price=price.fillna(method='ffill')
```

```python
djai['MONTH']=pd.DatetimeIndex(djai['Date']).month
djai['YEAR']=pd.DatetimeIndex(djai['Date']).year

djai.drop(['Open','High','Low','Adj Close','Volume','Date'],axis=1,inplace=True)

price=price.merge(djai,how='left')
price.rename(columns={'Close':'DJAI Close'},inplace=True)
```

```python
fed_spending.drop(['Air','Water','Railroad','Highway','Other'],axis=1,inplace=True)
price=price.merge(fed_spending,how='left')
```

```python
hai=hai.merge(inf,how='right')

hai['MONTH']=pd.DatetimeIndex(hai['Date']).month
hai['YEAR']=pd.DatetimeIndex(hai['Date']).year

hai.drop('Date',inplace=True,axis=1)

price=price.merge(hai,how='left')
```

```python
mort['YEAR']=pd.DatetimeIndex(mort['DATE']).year
mort['MONTH']=pd.DatetimeIndex(mort['DATE']).month

mort=mort.groupby(['YEAR','MONTH']).mean()
mort.reset_index(inplace=True)

price=price.merge(mort,how='left')
```

```python
permit['YEAR']=pd.DatetimeIndex(permit['DATE']).year
permit['MONTH']=pd.DatetimeIndex(permit['DATE']).month
```

```python
price=price.merge(permit,how='left')
price=price.merge(crime,how='left')
```

```python
price.drop(['MONTH','YEAR'],axis=1,inplace=True)

price['FDI(million)']=price['FDI(million)']/10
price['FDI(million)']=price['FDI(million)']/price['GDP(billion)']
price.rename(columns={'FDI(million)':'FDI % of GDP'},inplace=True)

price['Total(Fed Spending)']=price['Total(Fed Spending)']/10000000
price['Total(Fed Spending)']=price['Total(Fed Spending)']/price['GDP(billion)']
price.rename(columns={'Total(Fed Spending)':'Fed spending % of GDP'},inplace=True)

price['Inflation Value']=price['Inflation Value'].apply(lambda x:float(x[:4]))

price=price.iloc[:-7,:]
price=price.fillna(method='ffill')
```

```python
price.to_csv('combined_data.csv',index=False)
```
